```python
def compute_profit(line):
    x = (line['duration_sum'] - line['minutes_included']) * line['rub_per_minute'] # –í—ã—á–∏—Å–ª—è–µ–º —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω—ã—Ö –º–∏–Ω—É—Ç
    y = (line ['gb_used'] - line['mb_per_month_included'] / 1024)  * line['rub_per_gb'] # –ó–∞—Ç–µ–º —Ü–µ–Ω—É –ø–æ—Ç—Ä–∞—á–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–Ω–µ—Ç-—Ç—Ä–∞—Ñ–∏–∫–∞
    z = (line['sent_messages_monthly'] - line['messages_included']) * line['rub_per_message'] # –ò —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π
    
    result = x * (x > 0).astype(float) + y * (y > 0).astype(float) + z * (z > 0).astype(float) + line['rub_monthly_fee'] # –î–æ–º–Ω–æ–∂–∞–µ–º –∫–∞–∂–¥—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é –Ω–∞ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –¥–∞–µ—Ç 1 –µ—Å–ª–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è –±–æ–ª—å—à–µ –Ω—É–ª—è –∏ 0 - –µ—Å–ª–∏ –º–µ–Ω—å—à–µ –Ω—É–ª—è. –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –∑–∞–Ω—É–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ —Ç—Ä–∞—Ç –≤ –ø—Ä–µ–¥–µ–ª–∞—Ö —Ç–∞—Ä–∏—Ñ–Ω–æ–≥–æ –ø–∞–∫–µ—Ç–∞. 
    return result
```

```python
df_month_general_profit_final['month_profit'] = compute_profit(df_month_general_profit_final)
df_month_final_profit_cuted = df_month_general_profit_final[['user_id', 'month', 'month_profit']]
```

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: c—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å –æ—Ç —Ç–∞—Ä–∏—Ñ–∞ ultra –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–∏–±—ã–ª–∏ –æ—Ç —Ç–∞—Ä–∏—Ñ–∞ smart.

```python
alpha = .01 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤—ã–±–µ—Ä–µ–º –±–æ–ª—å—à–æ–π, —Ç.–∫. –µ—Å—Ç—å –≤—Å–µ –æ—Å–Ω–æ–≤–∞–Ω–∏—è –ø–æ–ª–∞–≥–∞—Ç—å, —á—Ç–æ —Ä–∞–∑–Ω–∏—Ü–∞ –≤ –ø—Ä–∏–±—ã–ª–∏ –≤ –ú–æ—Å–∫–≤–µ –∏ –≤ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–∞—Ö –∏–º–µ–µ—Ç—Å—è
# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –±–æ–ª—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

results = st.ttest_ind(avg_profit_Moscow, avg_profit_other_cities)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```

–î–ª—è —Ç–∞—Ä–∏—Ñ–∞ `smart`:

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É H‚ÇÄ: c—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ smart –∏–∑ –ú–æ—Å–∫–≤—ã —Ä–∞–≤–Ω–∞ —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–∏–±—ã–ª–∏ –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ smart –∏–∑ —Ä–µ–≥–∏–æ–Ω–æ–≤. –≠—Ç–æ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –≥–∏–ø–æ—Ç–µ–∑–∞.

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: c—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ smart –∏–∑ –ú–æ—Å–∫–≤—ã –Ω–µ —Ä–∞–≤–Ω–∞ —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–∏–±—ã–ª–∏ –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ smart –∏–∑ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –≥–æ—Ä–æ–¥–æ–≤.

```python
alpha = .01 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤—ã–±–µ—Ä–µ–º –±–æ–ª—å—à–æ–π, —Ç.–∫. –µ—Å—Ç—å –≤—Å–µ –æ—Å–Ω–æ–≤–∞–Ω–∏—è –ø–æ–ª–∞–≥–∞—Ç—å, —á—Ç–æ —Ä–∞–∑–Ω–∏—Ü–∞ –≤ –ø—Ä–∏–±—ã–ª–∏ –≤ –ú–æ—Å–∫–≤–µ –∏ –≤ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–∞—Ö –∏–º–µ–µ—Ç—Å—è
# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –±–æ–ª—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

data_Moscow = revenue.query('city == "–ú–æ—Å–∫–≤–∞"')
data_other_cities = revenue.query('city != "–ú–æ—Å–∫–≤–∞"')

data_Moscow_smart = data_Moscow[data_Moscow['tariff'] == 'smart']['month_profit']
data_other_cities_smart = data_other_cities[data_other_cities['tariff'] == 'smart']['month_profit']


results = st.ttest_ind(data_Moscow_smart, data_Moscow_smart)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```

–î–ª—è —Ç–∞—Ä–∏—Ñ–∞ `ultra`:

*–°—Ä–µ–¥–Ω—è—è –≤—ã—Ä—É—á–∫–∞ —Ç–∞—Ä–∏—Ñ–∞ ultra –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –∏–∑ –ú–æ—Å–∫–≤—ã –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –≤—ã—Ä—É—á–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ultra –∏–∑ –¥—Ä—É–≥–∏—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤.*

–°—Ñ–æ—Ä–º—É–ª–∏—Ä—É–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É H‚ÇÄ: c—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ ultra –∏–∑ –ú–æ—Å–∫–≤—ã —Ä–∞–≤–Ω–∞ —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–∏–±—ã–ª–∏ –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ ultra –∏–∑ —Ä–µ–≥–∏–æ–Ω–æ–≤. –≠—Ç–æ –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –≥–∏–ø–æ—Ç–µ–∑–∞.

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: c—Ä–µ–¥–Ω—è—è –ø—Ä–∏–±—ã–ª—å –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ ultra –∏–∑ –ú–æ—Å–∫–≤—ã –Ω–µ —Ä–∞–≤–Ω–∞ —Å—Ä–µ–¥–Ω–µ–π –ø—Ä–∏–±—ã–ª–∏ –æ—Ç –∫–ª–∏–µ–Ω—Ç–æ–≤ ultra –∏–∑ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –≥–æ—Ä–æ–¥–æ–≤.

```python
alpha = .01 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ –≤—ã–±–µ—Ä–µ–º –±–æ–ª—å—à–æ–π, —Ç.–∫. –µ—Å—Ç—å –≤—Å–µ –æ—Å–Ω–æ–≤–∞–Ω–∏—è –ø–æ–ª–∞–≥–∞—Ç—å, —á—Ç–æ —Ä–∞–∑–Ω–∏—Ü–∞ –≤ –ø—Ä–∏–±—ã–ª–∏ –≤ –ú–æ—Å–∫–≤–µ –∏ –≤ –¥—Ä—É–≥–∏—Ö –≥–æ—Ä–æ–¥–∞—Ö –∏–º–µ–µ—Ç—Å—è
# –µ—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –±–æ–ª—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–Ω–µ–º –≥–∏–ø–æ—Ç–µ–∑—É

data_Moscow = revenue.query('city == "–ú–æ—Å–∫–≤–∞"')
data_other_cities = revenue.query('city != "–ú–æ—Å–∫–≤–∞"')

data_Moscow_ultra = data_Moscow[data_Moscow['tariff'] == 'ultra']['month_profit']
data_other_cities_ultra = data_other_cities[data_other_cities['tariff'] == 'ultra']['month_profit']


results = st.ttest_ind(data_Moscow_ultra, data_Moscow_ultra)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
else:
    print('–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```
